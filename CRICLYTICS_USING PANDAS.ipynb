<h2><center><b>
ANALYSING THE Bowler_data.csv
</h2></center></b>

```
#To know the columns of the data
bow.columns
```

```
#To separate separate the Indian players from the data
#While analysing the WC_Players.csv we have extracted the Indian Players list to variable pl_Indians
# Bowler Data of Indian Players
cricket_bowler={}
for i in pl_Indians['ID']:
    for j in bow['Player_ID']:
        if i==j:
            a=bow.loc[bow['Player_ID'] == i]
            b=pl_Indians.loc[pl_Indians['ID']==j]
            Name=b['Player'].sum()
            mdns=a['Mdns'].sum()
            runs=a['Runs'].sum()
            wkts=a['Wkts'].sum()
            eco=a['Econ'].mean()
            sr=a['SR'].mean()
            cricket_bowler[i]=[Name,mdns,runs,wkts,eco,sr]
cricket_bowler=pd.DataFrame.from_dict(cricket_bowler).T
cricket_bowler=cricket_bowler.rename(columns={0:"Name",1:"MDNS",2:"Runs",3:"WKTS",4:"ECO",5:"SR"})

# Sorting the batsman data according to their Economy rate
cricket_bowler_LowEconomy=cricket_bowler.sort_values(['ECO'],ascending=True)
cricket_bowler_MaxWicket=cricket_bowler.sort_values(['WKTS'],ascending=False)
```

```
#The top 5 bowlers with low economy rate

Best_Bowler_A=cricket_bowler_LowEconomy.head(5)
Best_Bowler_A
```

```
#The Top 5 bowlers who take maximum wickets

Best_Bowler_B=cricket_bowler_MaxWicket.head(5)
Best_Bowler_B
```

```
#The Best Bowlers are 
Best_Bowler_A_Names=cricket_bowler_LowEconomy['Name'].head(5)
BBAL=set(Best_Bowler_A_Names)
Best_Bowler_B_Names=cricket_bowler_MaxWicket['Name'].head(5)
BBBL=set(Best_Bowler_B_Names)
Best_Bowlers=BBAL.union(BBBL)
Indian_Best_Bowlers=list(Best_Bowlers)
Indian_Best_Bowlers
```

<h3><b>INFERENCE:</h3></b><br>
<i>The best Bowlers with less economy rate and maximum wicket taking capability are as follows.
 </i>

1. Jasprit Bumrah
2. Kuldeep Yadav
3. Bhuvaneshwar Kumar
4. Ravindra Jadeja
5. Yuzvendra Chahal
6. Mohammed Shami

<h2><center><b>
ANALYSING THE Batsman_Data.csv
</h2></center></b>

```
#To know the columns of the data
bat.columns
```

```
#To separate separate the Indian players from the data
#While analysing the WC_Players.csv we have extracted the Indian Players list to variable pl_Indians
# Batsman Data of Indian Playersbat.columns
cricket_batsman={}
for i in pl_Indians['ID']:
    for j in bat['Player_ID']:
        if i==j:
            a=bat.loc[bat['Player_ID'] == i]
            b=pl_Indians.loc[pl_Indians['ID']==j]
            Name=b['Player'].sum()
            run=a['Runs'].mean()
            fours=a['4s'].sum()
            sixs=a['6s'].sum()
            sr=a['SR'].mean()
            cricket_batsman[i]=[Name,run,fours,sixs,sr]
cricket_batsman=pd.DataFrame.from_dict(cricket_batsman).T 
cricket_batsman=cricket_batsman.rename(columns={0:"Name",1:"Runs",2:"4s",3:"6s",4:"SR"})

# Sorting the batsman data according to the number of 6s and their strike rate
cricket_batsman=cricket_batsman.sort_values(['6s','SR'],ascending=False)

# Display the selected batsman
cricket_batsman
```

```
#Top 6 Batsman with good strike rate
Best_Batsman=cricket_batsman.head(10)
Best_Batsman=cricket_batsman['Name'].head(10)
Indian_Best_Batsman=list(Best_Batsman)
Indian_Best_Batsman
```
